```python
bat_scores = pd.read_csv(r'C:\Local Disk F\ML\Campus SuperBowl Challenge 2024\Data\batsman_stats.csv')
bowl_scores = pd.read_csv(r'C:\Local Disk F\ML\Campus SuperBowl Challenge 2024\Data\bowler_stats.csv')
```

```python
bat_scores = bat_scores.rename(columns = {'batsman_id': 'player_id'})
bowl_scores = bowl_scores.rename(columns = {'bowler_id': 'player_id'})
```

```python
performance_scores = pd.merge(bat_scores, bowl_scores, on='player_id', how='outer')

def calculate_total_score(row):
    bat_score = row['bat_performance_score']
    bowl_score = row['bowl_performance_score']
    
    if pd.notna(bat_score) and pd.notna(bowl_score):
        if bat_score > 5 and bowl_score < 5:
            return 0.7 * bat_score + 0.3 * bowl_score
        elif bat_score < 5 and bowl_score > 5:
            return 0.3 * bat_score + 0.7 * bowl_score
        elif bat_score > 5 and bowl_score > 5:
            return (bat_score + bowl_score) / 2
        else:
            return (bat_score + bowl_score) / 2
    elif pd.notna(bat_score):
        return bat_score * 0.75
    elif pd.notna(bowl_score):
        return bowl_score * 0.75
    else:
        return None

performance_scores['total_performance_score'] = performance_scores.apply(calculate_total_score, axis=1)
performance_scores = performance_scores.sort_values(by='total_performance_score', ascending=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import GradientBoostingClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from scipy.stats import mode
```

```python
model1 = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)
model2 = LGBMClassifier(n_estimators=100, learning_rate=0.1, max_depth=1, random_state=0)
model3= CatBoostClassifier(n_estimators=100, learning_rate=0.1, depth=1, random_state=0, verbose=0)
model4 = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=1, random_state=0, use_label_encoder=False, eval_metric='logloss')

model1.fit(X_train,y_train)
model2.fit(X_train,y_train)
model3.fit(X_train,y_train)
model4.fit(X_train,y_train)

```

```python
pred1=model1.predict_proba(X_test)
pred2=model2.predict_proba(X_test)
pred3=model3.predict_proba(X_test)
pred4=model4.predict_proba(X_test)

finalpred=(pred1*0.25+pred2*0.25+pred3*0.25+pred4*0.25)
```

```python
# Convert probabilities to class labels
final_predictions = np.argmax(finalpred, axis=1)

# Evaluate the ensemble model
accuracy = accuracy_score(y_test, final_predictions)
print(f"Ensemble Accuracy: {accuracy:.4f}")
```

```python
pred10 = np.argmax(pred1, axis=1)
pred20 = np.argmax(pred2, axis=1)
pred30 = np.argmax(pred3, axis=1)
pred40 = np.argmax(pred4, axis=1)

accuracy1 = accuracy_score(y_test, pred10)
accuracy2 = accuracy_score(y_test, pred20)
accuracy3 = accuracy_score(y_test, pred30)
accuracy4 = accuracy_score(y_test, pred40)

print(accuracy1, accuracy2, accuracy3, accuracy4)
```
